```python
# Import the necessary libraries
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, ENGLISH_STOP_WORDS
import pandas as pd
```

```python
# Load the dataset
newsgroups_data = fetch_20newsgroups(subset='train')
data = newsgroups_data['data']
```

```python
# https://towardsdatascience.com/basics-of-countvectorizer-e26677900f9c

# stop_words removes very common words like the or a as they dont provide any info
# as the corpus is in english, the usual english stopwords are a good start
# max_df in how many documents the term may occur at most to provide useful information e.g. Carmen in the same named Opera
# if the term occurs in at most 75% of the documents it may still provide useful information
# min_df e.g names which occur only in 1 document as they dont provide enough info on the whole dataset
# if the term occurs in less than 0,5% of the documents we considered it to not be useful
# max_features selects the x most common terms/words in the dataset 
# we chose 17 because it is nicely displayed

coun_vect = CountVectorizer(stop_words='english', max_df = 0.75, min_df = 0.005, max_features = 17)

count_matrix = coun_vect.fit_transform(data)

count_array = count_matrix.toarray()
df = pd.DataFrame(data=count_array,columns = sorted(coun_vect.vocabulary_.keys()))

# the label of the columns represents the vocabulary features whereas the rows represent the documents.
df
```

**Task Part 2: Exploratory Data Analysis of Text Data with Bag of Words Model**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = df.sum() # sums up the frequency counts across all documents
vocab = list(df1.keys())
frequency = list(df1[:])
 
# creating the bar plot to investigate the distribution 
# because they effect the outcome more we only display the most frequent words
fig = plt.figure(figsize = (10, 5))

plt.bar(vocab, frequency, color ='cyan',
        width = 0.4)
 
plt.xlabel("vocabulary")
plt.xticks(rotation = 45)
plt.ylabel("frequency")
plt.title("Frequencies of the most common words")
plt.show()

# when setting min_df = 1 the most frequent word 'ax' occurs about three times as often as the second frequent word 'edu'. 
# ax is already excluded because our min_df is at 0.5%
```

# Unusual stopwords

edu may come from educational journals 
nntp = network news transfer protocol because we are dealing with online sources 
com because we are dealing with online sources 
don may come from don't, might be a problem to exclude it, maybe better to use clitis and tranform n't to not 

```python
unusual_stopwords = ['edu', 'nntp', 'com', 'ax', 'don'] 

required_stopwords = ['an', 'a', 'the', 'and', 'it', 'for', 'or', 'but', 'in', 'my', 'your', 'our', 'their'] # might already be coveres by english stopwords

new_stopwords = unusual_stopwords + required_stopwords

stopwords = list(ENGLISH_STOP_WORDS.union(new_stopwords))
```

```python
# Re-run the count vectorizer with our stopwords
coun_vect = CountVectorizer(stop_words=stopwords, max_df = 0.75, min_df = 0.005)#, max_features = 17)

count_matrix = coun_vect.fit_transform(data)

count_array = count_matrix.toarray()
df = pd.DataFrame(data=count_array,columns = sorted(coun_vect.vocabulary_.keys()))

df
```

```python
df2 = df.sum() #count total frequencies of each word

vocab = list(df2.keys())
frequency = list(df2[:])

# get top 6 words by getting the indices of the highest frequencies
high_frequency_term = [] 
high_frequency_frequency = []
top_idx = np.argsort(frequency)[-6:]
for i in top_idx:
    high_frequency_term.append(vocab[i])
    high_frequency_frequency.append(frequency[i])
    
# creating the bar plot
fig = plt.figure(figsize = (10, 5))

plt.bar(high_frequency_term, high_frequency_frequency, color ='cyan',
        width = 0.4)
 
plt.xlabel("vocabulary")
plt.xticks(rotation = 45)
plt.ylabel("frequency")
plt.title("Frequencies of the most common words")
plt.show()

# get top 6 words by getting the indices of the lowest frequencies
low_frequency_term = [] 
low_frequency_frequency = []
low_idx = np.argsort(frequency)[:6]
for i in low_idx:
    low_frequency_term.append(vocab[i])
    low_frequency_frequency.append(frequency[i])

# creating the bar plot
fig = plt.figure(figsize = (10, 5))

plt.bar(low_frequency_term, low_frequency_frequency, color ='magenta',
        width = 0.4)
 
plt.xlabel("vocabulary")
plt.xticks(rotation = 45)
plt.ylabel("frequency")
plt.title("Frequencies of the least common words")
plt.show()
```

**Task Part 3: Text Classification with Bag of Words Model**

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
```

```python
# Load the dataset
newsgroups_data = fetch_20newsgroups()
data2 = newsgroups_data['data']
targets2 = newsgroups_data['target']
```

```python
# use new CountVectorizer on the whole fetch20_newsgroup data
coun_vect = CountVectorizer(stop_words = stopwords, max_df = .75, min_df = 0.005)
coun_vect.fit(data2)
```

```python
# split data into training and testing sets and their targets
train_set, test_set, train_targets, test_targets = train_test_split(data2, targets2, test_size = 0.2, random_state = 42)
```

```python
# Naive Bayes classifier
MNB_classifier = MultinomialNB()
MNB_classifier.fit(coun_vect.transform(train_set), train_targets)

# test the classifier
y_pred = MNB_classifier.predict(coun_vect.transform(test_set))

# Evaluation of classifier with appropriate metrics
print(classification_report(test_targets, y_pred, target_names=newsgroups_data['target_names']))
```

# Interpretation 

The accuracy is at 83% thus the result seems pretty good. Especially the classifier works well on rec.sport.hockey and sci.crypt. Yet it has a hard time classifying comp.sys.ibm.pc.hardware. Maybe because of some stopwords we excluded like nntp (we did not check that) 

# Parameters of CountVectorizer 

The baseline we compare our results to it the above.

Against our expectation the overall accuracy only dropped by 1% when excluding stopwords. 
Lowering the max_df to 50% or setting it to 100% didn't affect the accuracy. 
Setting min_df to 1 changed the accuracy to 87%; setting it to 3% the accuracy dropped to 8%. Probably because too many relevant terms were erased.
