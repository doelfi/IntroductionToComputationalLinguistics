```python
# Import the necessary libraries
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, ENGLISH_STOP_WORDS
import pandas as pd
```

```python
# Load the dataset
newsgroups_data = fetch_20newsgroups(subset='train')
data = newsgroups_data['data']
```

```python
# https://towardsdatascience.com/basics-of-countvectorizer-e26677900f9c


coun_vect = CountVectorizer(stop_words='english', max_df = 0.75, min_df = 1, max_features = 17)

# stop_words removes very common words like the or a as they dont provide any info
# max_df in how many documents the term may occur at most to provide useful information e.g. Carmen in the same named Opera
# min_df e.g names which occur only in 1 document as they dont provide enough info on the whole dataset
# max_freatures selects the x most common terms/words in the dataset

# 11314 rows × 129796 columns with stop_words = 'english'
# vmtl 11314 documents in train subset and vocab_size = 129796
count_matrix = coun_vect.fit_transform(data)

count_array = count_matrix.toarray()
df = pd.DataFrame(data=count_array,columns = coun_vect.vocabulary_.keys())

# print(count_matrix) sparse.csr_matrix spart Platz, weil ehh viele nullen

df 
```

**Task Part 2: Exploratory Data Analysis of Text Data with Bag of Words Model**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = df.sum()
vocab = list(df1.keys())
#print(vocab)
frequency = list(df1[:])

fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(vocab, frequency, color ='cyan',
        width = 0.4)
 
plt.xlabel("vocabulary")
plt.ylabel("frequency")
plt.title("Frequencies of the most common words")
plt.show()
```

```python
unusual_stopwords = ['edu', 'nntp', 'com', 'ax', 'don'] 
required_stopwords = ['an', 'a', 'the', 'and', 'it', 'for', 'or', 'but', 'in', 'my', 'your', 'our', 'their'] # might already be coveres by english stopwords

new_stopwords = unusual_stopwords + required_stopwords

stopwords = list(ENGLISH_STOP_WORDS.union(new_stopwords))
```

```python
coun_vect = CountVectorizer(stop_words=stopwords, max_df = 0.75, min_df = 1)#, max_features = 17)

count_matrix = coun_vect.fit_transform(data)

count_array = count_matrix.toarray()
df = pd.DataFrame(data=count_array,columns = coun_vect.vocabulary_.keys())

df
```

```python
df2 = df.sum() #count total frequencies ofo each word

vocab = list(df2.keys())
frequency = list(df2[:])

# get top 6 words by getting the indices of the highest frequencies
top_idx = np.argsort(frequency)[-6:]
top_values = [(vocab[i],frequency[i]) for i in top_idx]

# get top 6 words by getting the indices of the lowest frequencies
low_idx = np.argsort(frequency)[:6]
low_values = [(vocab[i],frequency[i]) for i in low_idx]

print(top_values)
print(low_values)
```
